# 1. Liste as diferenças entre uma lista, um dicionário, um array numpy e uma series pandas?


Uma lista é uma sequência ordenada de valores que são indentificadas por índices, sendo melhor indicado usar para manipular coleções ordenadas. Um dicionário é um tipo de mapeamento associativo entre chave e valor, que é mais usado para busca de elementos a partir de um parâmetro que aponte para um elemento. Um array em NumPy é uma tabela com diversas dimensões de diversos elementos de um mesmo tipo, onde das dimensões são chamadas de axes, normalmente usado para manipulação algébrica dos dados. Uma serie pandas são arranjos ou matrizes unidimensionais capazes de armazenar dados de qualquer tipo.

# 8. Como alterar a ordem das colunas de um dataframe? Para isso, respondam as três questões:



# (a) Em df, troque as colunas ’a’ e ’c’.

```python
import pandas as pd

dados = [[1,2,3], [4,5,6], [7,8,9]]
colunas = ["a", "b", "c"]
df = pd.DataFrame(data = dados, columns = colunas)
df.head()
```

Alteração da ordem das colunas 'a' e 'c'

```python
dados = [[1,2,3], [4,5,6], [7,8,9]]
colunas = ["c", "b", "a"]
df = pd.DataFrame(data = dados, columns = colunas)
df.head()
```

# (c) Classifique as colunas em ordem alfabética inversa, ou seja, colume ’e’ primeiro até a coluna ’a’ last.

```python
dados = [[1,2,3,9,8,17], [4,5,1,4,5,6], [7,95,8,2,8,9], [11,15,100,1,19,3], [1,0,4,5,15,36], [144,1,77,18,29]]
colunas = ["a", "b", "c", "d", "e", "f"]
df = pd.DataFrame(data = dados, columns = colunas)
df
```

Inversão das colunas 

```python
dados = [[1,2,3,9,8,17][::-1], [4,5,1,4,5,6][::-1], [7,95,8,2,8,9][::-1], [11,15,100,1,19,3][::-1], [1,0,4,5,15,36][::-1], [144,1,77,18,29][::-1]]
colunas = ["a", "b", "c", "d", "e", "f"][::-1]
df = pd.DataFrame(data = dados, columns = colunas)
df
```
