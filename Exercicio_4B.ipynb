```python
#Relação de pesquisadores 2019
#Link1

url = 'https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36821'
RelacaoPesquisadores = pd.read_csv(url, encoding = 'cp1252', sep = ',')
RelacaoPesquisadores.head()
```

```python
#Relação de projetos de pesquisa realizados 2019
#Link1

url = 'https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36822'
RelacaoProjetos = pd.read_csv(url, encoding = 'cp1252', sep = ',')
RelacaoProjetos.head()
```

```python
#Relação de projetos de pesquisa realizados 2019
#Link2

url = 'https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36482'
ProjetosDeExtensao = pd.read_csv(url, encoding = 'cp1252', sep = ',')
ProjetosDeExtensao.head()
```

Junção dos arquivos referentes aos links 1, 2, respectivamente, que contemplam os projetos. 

```python
NovaRelacaoProjetos1 = RelacaoProjetos
NovaRelacaoProjetos2 = ProjetosDeExtensao
NovaRelacaoProjetos1.dropna(how='any')
NovaRelacaoProjetos2.dropna(how='any')
novoDF = pd.concat([NovaRelacaoProjetos1, NovaRelacaoProjetos2], axis = 0, ignore_index = True)
novoDF
```

```python
novoDF_BACKUP = novoDF
novoDF_BACKUP.to_csv("novoDF_BACKUP.csv")
novoDF_BACKUP
```

É necessário responder a pergunta: o que é JSON e quais as vantagens dele sobre o CSV?

Arquivos com extensão .csv são usado de modo geral apenas para a exportação e importação dos dados. Arquivos com extensão .json, além de serem usados para importação e exportação, são usados em solução de troca de dados na comunição em rede - serviços WEB - envolvemdo clientes e aplicações de maneira bidirecional.
